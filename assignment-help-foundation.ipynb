### Principal Component Analysis

```python
#Importing the PCA module
from sklearn.decomposition import PCA
pca = PCA(svd_solver='randomized', random_state=35)
```

```python
#Doing the PCA on our data
pca.fit(data)
```

```python
pca.components_
```

```python
pca_df = pd.DataFrame({'Feature':data_col,'PC1':pca.components_[0],'PC2':pca.components_[1]})
pca_df
```

```python
%matplotlib inline
fig = plt.figure(figsize = (8,8))
plt.scatter(pca_df.PC1, pca_df.PC2)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
for i, txt in enumerate(pca_df.Feature):
    plt.annotate(txt, (pca_df.PC1[i],pca_df.PC2[i]))
plt.tight_layout()
plt.show()
```

We see that the first component is in the direction where the 'income','g variables are heavy

<b> Screeplot to assess the number of needed principal components

```python
pca.explained_variance_ratio_
```

```python
#Making the screeplot - plotting the cumulative variance against the number of components
%matplotlib inline
fig = plt.figure(figsize = (8,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

We can see that 5 components can describe about 95% variance.

### Clustering

<b>Hopkins Statistics

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
hopkins(data_after_pca)
```

the hopkins statistics gives a value greater than 0.74 i.e it has a high tendency to cluster.

Therefore we will proceed with clustering methodology.

### K-Means algorithm

<b>Choosing the number of clusters K in advance
  <p>  1.Elbow Method
  <p> Calculate sum of squared distances    

```python
# sum of squared distances
# To perform KMeans clustering 
from sklearn.cluster import KMeans

ssd = []
for num_clusters in list(range(1,21)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=50)
    model_clus.fit(data_after_pca)
    ssd.append(model_clus.inertia_)

plt.plot(ssd)
```

<b>2. Average Silhouette Method

```python
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k).fit(data_after_pca)
    sse_.append([k, silhouette_score(data_after_pca, kmeans.labels_)])
```

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

From the above two methods, we conclude to take number of clusters as 3.

```python
# Kmeans with K=3
model1_kmeans = KMeans(n_clusters = 3, max_iter=50,random_state=100)
model1_kmeans.fit(data_after_pca)
```
